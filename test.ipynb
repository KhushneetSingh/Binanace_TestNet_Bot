```python
from binance.client import Client
import time

# Tested this with Binance Spot Test Network API
# Get your API keys from https://testnet.binance.vision/

API_KEY = 'Put your API key here'
API_SECRET = 'Put your API secret here'
```

```python
client = Client(API_KEY, API_SECRET, testnet=True)
```

```python
client.get_account()
```

```python
symbol = 'BTCUSDT'
buy_price = 30000.0
sell_price = 35000.0
quantity = 0.001
```

```python
def get_current_price(symbol):
    ticker = client.get_symbol_ticker(symbol=symbol)
    return float(ticker['price'])
```

```python
get_current_price(symbol)
```

```python
def place_buy_order(symbol, quantity):
    order = client.order_market_buy(symbol=symbol, quantity=quantity)
    print(f"Buy order placed: {order}")
```

```python
place_buy_order(symbol, quantity)
```

```python
def place_sell_order(symbol, quantity):
    order = client.order_market_sell(symbol=symbol, quantity=quantity)
    print(f"Sell order placed: {order}")
```

```python
place_sell_order(symbol, quantity)
```

```python
def trading_bot():
    while True:
        current_price = get_current_price(symbol)
        print(f"Current price of {symbol}: {current_price}")

        if current_price <= buy_price:
            place_buy_order(symbol, quantity)
        elif current_price >= sell_price:
            place_sell_order(symbol, quantity)

time.sleep(3)  # Wait for 1 minute before checking the price again
```

```python
if __name__ == "__main__":
    trading_bot()
```
